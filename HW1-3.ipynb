## Part 3: Write a query function (10 pts)

Write a function called `query_climate()` which accepts five arguments:

`df`, the file name for the dataframe

`country`, a string giving the name of a country for which data should be returned.

`year_begin` and `year_end`, two integers giving the earliest and latest years for which should be returned.

`month`, an integer giving the month of the year for which should be returned.

The return value of `query_climate()` is a Pandas dataframe of temperature readings for the specified country, in the specified date range, in the specified month of the year. This dataframe should have the following columns, **in this order**:

`NAME`: The station name.

`LATITUDE`: The latitude of the station.

`LONGITUDE`: The longitude of the station.

`Country`: The name of the country in which the station is located.

`Year`: The year in which the reading was taken.

`Month`: The month in which the reading was taken.

`Temp`: The temperature at the specified station during the specified year and month. 


**Grading is based on unseen test examples. A test example is given below.**

```python
def query_climate(df, country, year_begin, year_end, month):
    df["Month"] = month
    selectedDf = df[(df['Name'] == country)]
    selectedDf = selectedDf[(selectedDf['Year'] >= year_begin) & (selectedDf['Year'] <= year_end)]
    selectedDf = selectedDf.loc[:, ['NAME', 'LATITUDE', 'LONGITUDE', 'Name', 'Year', 'Month', f'VALUE{month}']] 
    selectedDf = selectedDf.rename(columns = {f'VALUE{month}' : "Temp"})
    selectedDf['Temp'] = selectedDf['Temp'] / 100

    return selectedDf
```

## Part 4: Global warming ( 10 pts )

In this part, the goal is to create a bar plot which is used to compare average temperatures of decadeds 1901-1910 and 2011-2020.

The bar plot looks like the following:



In above plot, I select three different countries and compute the average temperatures of year range 1901-1910 and 2011-2020 in January. Then, I use (two) barplot(s) to visualize the average temperature. 

**You should do more than my sample plot:**

Select 4 different countries and 6 different months, then compute two average temperatures for each country in each certain month. Then visualize 6 different plots (included in a single plot using `subplot` command). Please add necessary labels, titles, and legends to make your plot clear.

You should write a function or class to simplify your code if necessary.

What is your conclusion?

```python
def plot_average_temperature_comparison(df, selected_countries, figsize=(18, 10)): #creating a function that takes in the data frame, which countries are to be selected for the plot and specifies the figure size if not provided
    fig, axes = plt.subplots(2, 3, figsize=figsize)

    for i, ax in enumerate(axes.flatten(), start=1):
        month_col = f'VALUE{i}'
        
        #Filter data for the first year range
        average_temp_1901_1910 = df[df['Year'].between(1901, 1910) & df['Name'].isin(selected_countries)][['Name', month_col]].groupby('Name').mean().reset_index()
        average_temp_1901_1910['Year Range'] = '1901-1910'

        #Filter data for the second year range
        average_temp_2011_2020 = df[df['Year'].between(2011, 2020) & df['Name'].isin(selected_countries)][['Name', month_col]].groupby('Name').mean().reset_index()
        average_temp_2011_2020['Year Range'] = '2011-2020'

        #Concatenate the two average temperature DataFrames
        average_temps = pd.concat([average_temp_1901_1910, average_temp_2011_2020])

        #Sort the DataFrame by 'Year Range' so that 2011-2020 is plotted on top
        average_temps_sorted = average_temps.sort_values(by='Year Range', ascending=False)

        #Plot 
        sns.barplot(data=average_temps_sorted, x='Name', y=month_col, hue='Year Range', dodge=False, palette='coolwarm', ax=ax)
        ax.set_xlabel('Country')
        ax.set_ylabel('Average Temperature (°C)')
        ax.set_title(f'Average Temperature Comparison for Month {i}')
        ax.legend(title='Year Range')

    plt.tight_layout()
    plt.show()

plot_average_temperature_comparison(df, selected_countries=['India', 'Japan', 'Algeria', 'Brazil'])
```

Conclusion: Surprisingly, some months have been more affected by global warming in certain countries than others. For example, this is demonstrated by the blue box in Japan for Month 4 being much shorter than the one in Japan for Month 5. Also, this plot visually demonstrates the fluctuations in temperatures over the 6 months for all countries.
